```python
def remove_space(list_headers,charact): #charact should be: charact=[' - ',' ']
    new_headers=list()
    for header in list_headers:
        for char in charact:      
            if char in header:
                header=header.replace(char,'_')
            header=header
        new_headers.append(header)
    return new_headers
```

```python
def remove_sp_char(headers,chars):
    new_headers=list()
    for header in headers:    
        for char in chars:
            if char=='-' or char=='/':
                header=header.replace(char,'_')
            if char in header:
                header=header.replace(char,'')
            
            header=header   
        new_headers.append(header)
    return new_headers    
```

```python
headers=remove_space(universities_df.columns,[' - ',' '])
```

```python
headers=remove_sp_char(headers,["'",',',':','-','/'])
```

```python
list_new_header=list()

for header in headers:
    header=header.casefold()   # All capitalized letters are changed.
            
    list_new_header.append(header)
universities_df.columns=list_new_header
```

## 3. Exploratory Data Analysis (EDA) and Visualization

```python
# Universities received the highest number of applications (20)
high_app_df=universities_df[['name','applicants_total']].sort_values('applicants_total',ascending=False).head(20)
print(high_app_df)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='applicants_total',y='name',data=high_app_df)
plt.title('Top 20 Indian Universities with the Most Applications')
plt.xlabel('Number of applications')
plt.ylabel('')
```

```python
# Universities received the highest number of admission (20)
high_add_df=universities_df[['name','admissions_total']].sort_values('admissions_total',ascending=False).head(20)
print(high_add_df)
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='admissions_total',y='name',data=high_add_df)
plt.title('Top 20 Indian Universities with the Most Admissions')
plt.xlabel('Number of admissions')
plt.ylabel('')
```

```python
#APPLICATIONS VS ADMISSIONS
plt.figure(figsize=(16,6))
plt.title('APPLICATIONS VS ADMISSIONS')
sns.scatterplot(y=universities_df.admissions_total,x=universities_df.applicants_total,hue=universities_df.control_of_institution)
plt.ylabel('Number of Admissions')
plt.xlabel('Number of Applications')
plt.grid()
```

```python
#spliting the number of applications according to the type of control: private or public.
uni_private_df=universities_df[universities_df.control_of_institution=='Private not-for-profit']
uni_private_df=uni_private_df[uni_private_df.applicants_total.notnull()]

uni_public_df=universities_df[universities_df.control_of_institution=='Public']
uni_public_df=uni_public_df[uni_public_df.applicants_total.notnull()]
```

```python
plt.figure(figsize=(16,7))
plt.hist([uni_public_df.applicants_total,uni_private_df.applicants_total],stacked=False,bins=25)
plt.axis([0,30000,0,200])
plt.title('Distribution of Applications')
plt.xlabel('Number of Applications')
plt.ylabel('Universities')
plt.legend(['Public universities. ({})'.format(len(uni_public_df)),'Private universities. ({})'.format(len(uni_private_df))]);
```
